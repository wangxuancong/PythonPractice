```python
import pandas as pd 
from pandas import Series
```

```python
file = 'BTC.csv'
data = pd.read_csv(file)
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date',inplace=True)
new_index = [i for i in range(data.shape[0])]
data.index = new_index
data.reindex
```

```python
data.head()
```

```python
data['shift1'] = data['Close'].shift(1)
```

```python
data.head()
```

```python
data['trend']= (data['Close']-data['shift1'])/data['shift1']
data.head()
```
